```python
def confusion_matrix(y,y_pred):
    count_correct_points = 0
    for i in range(y.shape[0]):
        if(y[i]==y_pred[i]):
            count_correct_points = count_correct_points + 1
    misclassified_count = y.shape[0] - count_correct_points
    return count_correct_points,misclassified_count
```

```python
from sklearn.svm import SVC
rbf_svc = SVC(kernel='rbf')
```

```python
rbf_svc = rbf_svc.fit(train_data[:,0:1],train_data[:,2])
```

```python
y_pred = rbf_svc.predict(test_data[:,0:1])
```

```python
correct,incorrect = confusion_matrix(test_data[:,2],y_pred)
```
